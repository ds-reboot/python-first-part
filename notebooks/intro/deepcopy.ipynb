```python
def check_exception(d):
    try:
        x = 10/d
    except ZeroDivisionError as e:
        print(e, 'Так не надо делить')
        x = 10/(d + 1)
    else:
        print('Отработало без ошибок')
    finally:
        print('Строка выводится вне зависимости от ошибок')
    return x
```

```python
check_exception(10)
```
