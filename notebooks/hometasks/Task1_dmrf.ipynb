```python
import requests
```

## Список объектов

```python
# Для скачивания данных по

offset_ = 0
limit_ = 100
objects_list = []
while True:
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = requests.get(url)
    request_list = res.json().get('data').get('list')
    if len(request_list) == 0:
        break
    objects_list.extend(request_list)
    offset_ += 100
    print(offset_)
```

```python
import pandas as pd
from pandas import DataFrame as df
```

```python
buildings = df.from_dict(objects_list, orient='columns')
buildings
```

# Получение таблицы девелоперов

```python
developers = pd.json_normalize(buildings['developer']).drop_duplicates()
developers
```

# Замена столбца в исходном датафрейме с данными застройщиков на идентификаторы

```python
buildings['devId'] = buildings['developer'].apply(lambda x: x['devId'])
buildings = buildings.drop('developer', axis=1)
buildings
```

# Сохранение в БД, Excel, pickle

```python
import sqlite3
conn = sqlite3.connect('buildings_database')
buildings.to_sql('buildings', conn, if_exists='replace', index = False)
developers.to_sql('developers', conn, if_exists='replace', index = False)
conn.commit()
conn.close()

buildings.to_excel('buildings.xlsx', index=False)
developers.to_excel('developers.xlsx', index=False)

buildings.to_pickle('buildings.pkl')
developers.to_pickle('developers.pkl')
```

# Загрузка фотографий

```python
def get_photo(obj_id):
    from urllib import request
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{obj_id}'
    res = requests.get(url)
    photo_data = res.json().get('data').get('photoRenderDTO')
    for photo in photo_data:
        if photo.get('objRenderPhotoUrl'):
            name = photo.get('objRenderPhotoNm')
            print(name)
            from pathlib import Path
            Path(f'photos/{obj_id}').mkdir(parents=True, exist_ok=True)
            response = request.urlretrieve(photo.get('objRenderPhotoUrl'), f'photos/{obj_id}/{name}')
```
